```python
def oauth_cb(oauth_config):
    auth_token, expiry_ms = MSKAuthTokenProvider.generate_auth_token("ap-northeast-2")
    # Note that this library expects oauth_cb to return expiry time in seconds since epoch, while the token generator returns expiry in ms
    return auth_token, expiry_ms/1000
```

```python
from json import loads


consumer = KafkaConsumer(
    bootstrap_servers="b-2-public.dp.ugprbm.c3.kafka.ap-northeast-2.amazonaws.com:9198",
    client_id=socket.gethostname(),
    security_protocol="SASL_SSL",
    sasl_mechanism="OAUTHBEARER",
    sasl_oauth_token_provider=tp,
    group_id="mygroup",
    auto_offset_reset="earliest",
    value_deserializer=lambda x: loads(x.decode("utf-8"))
)
```

```python
while True:
    msg = consumer.poll(5)
    print('Received message: {}'.format(msg.get))

consumer.close()
```
